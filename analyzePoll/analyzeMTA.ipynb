This file will check how much financial stress the mta causes / reliability of the mta 


```python
import pandas as pd 

# find the median of this column "How many MetroCard swipes (excluding transfers) do you use on a typical day?"

# CunyMetroCard191.xlsx is the file name
df = pd.read_excel('CunyMetroCard191.xlsx')
medianMetroCardSwipes = df['How many MetroCard swipes (excluding transfers) do you use on a typical day?'].median()
meanMetroCardSwipes = df['How many MetroCard swipes (excluding transfers) do you use on a typical day?'].mean()

print(medianMetroCardSwipes)
print(round(meanMetroCardSwipes)) # round to the nearest integer
```

```python
# how many metrocard swipes are used for college
medianMetroCardSwipesForCollege = df['How many of those swipes are used for commuting to and from college?'].median()
meanMetroCardSwipesForCollege = df['How many of those swipes are used for commuting to and from college?'].mean()

print(medianMetroCardSwipesForCollege)
print(round(meanMetroCardSwipesForCollege)) # round to the nearest integer
```

```python
# how much financial stress does the metrocard cause (0 - 10 scale -> 0 being no stress, 5 -> moderate stress, 10 -> extreme stress)
meanFinancialStress = df['On a scale of 1-10, how much financial stress does transportation cause you? (Linear scale: 1 = No stress, 10 = Extreme stress)'].mean()
print("MEAN FINANCIAL STRESS ",  meanFinancialStress) # round to the nearest integer

# how reliable do you find the mta (1-> unreliable, 5 -> somewhat reliable, 10-> very reliable)
meanReliability = df['On a scale of 1-10, how reliable do you find the MTA? (Linear scale: 1 = Very unreliable, 10 = Very reliable)'].mean()
print("MEAN MTA RELIABILITY ", meanReliability) # round to the nearest integer
```

```python
# generate a correlation scatter plot with y axis (How many MetroCard swipes (excluding transfers) do you use on a typical day?) and x axis (On a scale of 1-10, how much financial stress does transportation cause you? (Linear scale: 1 = No stress, 10 = Extreme stress)))

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['On a scale of 1-10, how much financial stress does transportation cause you? (Linear scale: 1 = No stress, 10 = Extreme stress)'], 
                y=df['How many MetroCard swipes (excluding transfers) do you use on a typical day?'])


plt.title('Correlation between MTA Reliability and MetroCard Swipes')

plt.xlabel('Financial Stress (1-10)')
plt.ylabel('MetroCard Swipes (excluding transfers)')
plt.grid()
plt.show()
```
